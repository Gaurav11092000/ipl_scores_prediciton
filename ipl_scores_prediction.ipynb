```python
from datetime import datetime
df['date'] = df['date'].apply(lambda x: datetime.strptime(x , '%Y-%m-%d'))
```

```python
dummy = pd.get_dummies(df , columns = ['bat_team' , 'bowl_team'])
```

```python
dummy['date'] = df['date']
```

```python
columns = ['date' ,'bat_team_Chennai Super Kings', 'bat_team_Delhi Daredevils',
       'bat_team_Kings XI Punjab', 'bat_team_Kolkata Knight Riders',
       'bat_team_Mumbai Indians', 'bat_team_Rajasthan Royals',
       'bat_team_Royal Challengers Bangalore', 'bat_team_Sunrisers Hyderabad',
       'bowl_team_Chennai Super Kings', 'bowl_team_Delhi Daredevils',
       'bowl_team_Kings XI Punjab', 'bowl_team_Kolkata Knight Riders',
       'bowl_team_Mumbai Indians', 'bowl_team_Rajasthan Royals',
       'bowl_team_Royal Challengers Bangalore',
       'bowl_team_Sunrisers Hyderabad', 'overs', 'runs_last_5',
       'wickets_last_5', 'runs', 'wickets', 'total']
```

```python
dummy = dummy[columns]
```

```python
dummy.head()
```

```python
X_train = dummy.drop('total' , axis = 1)[dummy['date'].dt.year <= 2016]
X_test = dummy.drop('total' , axis = 1)[dummy['date'].dt.year >= 2017]
```

```python
Y_train = dummy[dummy['date'].dt.year <= 2016]['total'].values
y_test =  dummy[dummy['date'].dt.year >= 2017]['total'].values
```

X_train.drop('date', axis = True , inplace = True)


```python
X_test.drop('date' , axis = True , inplace = True)
```

```python
from sklearn.linear_model import Lasso
from sklearn.model_selection import GridSearchCV
```

```python
lasso = Lasso()
parameters = {'alpha' : [1e-15 , 1e-10 , 1e-8 , 1e-3 , 1e-2 , 1 , 5 ,10 , 20 , 30 , 35 , 40]}
lasso_regressor = GridSearchCV(lasso , parameters , scoring = 'neg_mean_squared_error' , cv= 5)

lasso_regressor.fit(X_train , Y_train)
print(lasso_regressor.best_params_)
print(lasso_regressor.best_score_)
```

```python
prediction = lasso_regressor.predict(X_test)
```

```python

import seaborn as sns
sns.distplot(y_test-prediction)
```

```python
from sklearn import metrics
import numpy as np
print('MAE:', metrics.mean_absolute_error(y_test , prediction))
print('MSE:' , metrics.mean_squared_error(y_test , prediction))
print('RMSE:' , np.sqrt(metrics.mean_squared_error(y_test , prediction)))
```
